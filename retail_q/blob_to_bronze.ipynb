```python
# Auto Loader streaming ingestion from Volume to Bronze table


# Read CSV files using Auto Loader (cloudFiles)
df = (spark.readStream
  .format("cloudFiles")
  .option("cloudFiles.format", "csv")
  .option("cloudFiles.schemaLocation", "/Volumes/retail_q/volumes/blob_source/_schemas/transaction")
  .option("header", "true")
  .option("inferSchema", "true")
  .load('/Volumes/retail_q/volumes/blob_source/transaction_source/')
)

# Write stream to target bronze table
query = (df.writeStream
  .option("checkpointLocation", "/Volumes/retail_q/volumes/blob_source/_checkpoints/retail_q_blob_bronze_transaction")
  .trigger(availableNow=True)
  .toTable("retail_q.blob_bronze.transaction")
)

query.awaitTermination()
```
