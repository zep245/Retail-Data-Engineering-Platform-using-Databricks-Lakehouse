```sql
CREATE VIEW retail_q.retail_semantic.retail_metrics
WITH METRICS
LANGUAGE YAML
AS $$
version: 1.1
source: retail_q.retail_gold.fact_sales
comment: Retail sales metrics for analyzing revenue, quantity, discounts, and transaction patterns
joins:
  - name: products
    source: retail_q.retail_gold.dim_product
    on: source.product_id = products.product_id
  - name: calendar
    source: retail_q.retail_gold.dim_calendar
    on: source.transaction_date = calendar.date
dimensions:
  - name: Transaction Date
    expr: calendar.date
    comment: Date when the transaction occurred
    display_name: Transaction Date
    format:
      type: date
      date_format: year_month_day
  - name: Product Name
    expr: products.product_name
    comment: Name of the product sold
    display_name: Product Name
  - name: Product Segment
    expr: products.product_segment
    comment: Market segment of the product
    display_name: Product Segment
  - name: Sales Channel
    expr: source.sales_channel
    comment: Channel through which the sale was made
    display_name: Sales Channel
  - name: Payment Mode
    expr: source.payment_mode
    comment: Method of payment used for the transaction
    display_name: Payment Mode
measures:
  - name: Total Revenue
    expr: SUM(source.amount)
    comment: Total revenue from all transactions
    display_name: Total Revenue
    format:
      type: currency
      currency_code: USD
      decimal_places:
        type: exact
        places: 2
    synonyms:
      - revenue
      - sales
  - name: Total Quantity
    expr: SUM(source.quantity)
    comment: Total quantity of products sold
    display_name: Total Quantity
    format:
      type: number
      decimal_places:
        type: exact
        places: 0
    synonyms:
      - quantity sold
      - units sold
  - name: Total Discount
    expr: SUM(source.discount_amount)
    comment: Total discount amount applied across all transactions
    display_name: Total Discount
    format:
      type: currency
      currency_code: USD
      decimal_places:
        type: exact
        places: 2
  - name: Transaction Count
    expr: COUNT(1)
    comment: Number of transactions
    display_name: Transaction Count
    format:
      type: number
      decimal_places:
        type: exact
        places: 0
    synonyms:
      - order count
      - number of orders
  - name: Average Order Value
    expr: MEASURE(`Total Revenue`) / MEASURE(`Transaction Count`)
    comment: Average revenue per transaction
    display_name: Average Order Value
    format:
      type: currency
      currency_code: USD
      decimal_places:
        type: exact
        places: 2
    synonyms:
      - AOV
      - avg order size
$$
```
